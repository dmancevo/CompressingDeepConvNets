```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

%matplotlib inline

sns.set_palette(sns.color_palette("Blues"))
```

```python
with open('cifar-10_data.txt', 'r') as f:
    cifar_10 = [float(d) for d in re.findall(r'(\d+\.?\d*)\%',f.read())]
      
cifar_10.reverse()
cifar_10 = pd.Series(cifar_10)
```

```python
titles = [
    "Multi-Column Deep NN's (2012).",
    "Stacked What-Where Auto-encoders (2015).",
    "Fractional Max-Pooling (2015).",
]

n_params = [
    2.,
    11.8,
    58.5
]
```

```python
sns.set(style="white", font_scale=2.0)

plt.bar([7,23,39], n_params, width=11, tick_label=titles, color=sns.color_palette("Blues"))
ax = plt.gca()
ax.set_ylabel("No. Parameters (millions)")
ax = ax.twinx()
ax = cifar_10.plot(marker='o', color='darkslateblue', label='CIFAR-10', legend=True, figsize=(20, 10))#, ax=ax)
ax.set_ylabel("Model accuracy (%)")

#13, 88.79
#30, 92.23
#46, 96.53

sns.set(style="white", font_scale=1.3)

ax.annotate("Multi-Column Deep NN's (2012).", xy=(13, 88.79),  xytext=(5, 90),arrowprops=dict(arrowstyle="->"))
ax.annotate("Stacked What-Where Auto-encoders (2015).", xy=(30, 92.23),  xytext=(23, 89),arrowprops=dict(arrowstyle="->"))
ax.annotate("Fractional Max-Pooling (2015).", xy=(46, 96.53),  xytext=(35, 96.8),arrowprops=dict(arrowstyle="->"))
```
