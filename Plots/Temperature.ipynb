```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.rcParams['figure.figsize'] = (7,14)

sns.set(style="whitegrid", font_scale=2.3)
```

```python
import pickle as pkl

with open('/home/diegom/notebooks/Data/cifar10/vgg16_logits.pkl','rb') as f:
    logits = pkl.load(f)

temp = lambda x, T: np.exp(x/T) / np.sum(np.exp(x/T))
```

```python
zero_logits = (logits - np.mean(logits, axis=0))
```

```python
sns.barplot(range(10), logits[0], palette=sns.color_palette("GnBu_d"))
```

```python
plt.subplots(figsize=(20,7))
for i,t in [(1,1),(2,3),(3,7)]:
    plt.subplot(1,3,i)
    ax = sns.barplot(range(10), temp(logits[0], t), palette=sns.color_palette("GnBu_d"))
    ax.set_ylim([0,1.])
    ax.annotate('T={0}'.format(t), (0.02,0.9), xycoords='axes fraction', fontsize=27)
#     print dir(ax)
```
