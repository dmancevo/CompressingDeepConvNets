```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.rcParams['figure.figsize'] = (12, 6)

sns.set(style='white', font_scale=1.7)

palette = sns.color_palette("Blues")
```

```python
plt.axes()

ax = plt.gca()

rectangle = plt.Rectangle((27.5, 0), 20, 10, fc=palette[0])
ax.add_patch(rectangle)

rectangle = plt.Rectangle((2.5, 0), 20, 10, fc=palette[2])
ax.add_patch(rectangle)

rectangle = plt.Rectangle((15, 15), 20, 10, visible=False)
ax.add_patch(rectangle)

rectangle = plt.Arrow(12.5, 11, 3, 7)
ax.add_patch(rectangle)

rectangle = plt.Arrow(37.5, 11, -3, 7)
ax.add_patch(rectangle)

plt.axis('scaled')
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ax.annotate("Teacher Model ($\hat{z}_t$)", xy=(3, 5))
ax.annotate("Student Model ($\hat{z}_s)$", xy=(28, 5))


ax.annotate("$\mathcal{L}_{reg} = ||\hat{z}_t - \hat{z}_s||^2_2$", xy=(18, 20))
```
